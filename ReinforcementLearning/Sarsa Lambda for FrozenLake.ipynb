```python
import gym
import numpy as np
```

# Game

```python
env = gym.make("FrozenLake-v0")
```

# Model

### Constants

```python
ACTIONS = [] # Action space
STATES = [] # State space
N_ACTIONS = env.action_space.n
N_STATES = env.observation_space.n
epsilon = 0.9 # Expoit v.s. Explore, this number will be updated later as the learning progress
gamma = 0.95 # Decay, same as discount for time value of money
LR = 0.8 # Learning Rate
N_GAMES = 100 #Training length
N_STEPS = 99 # How many updates each game have
lambda_= 0.9 # Eligibility Race Decay
```

### Q Table

```python
Q = np.zeros((N_STATES, N_ACTIONS)) + 0.1
```

```python
E = np.zeros((N_STATES, N_ACTIONS))  + 0.1# Eligibility Trace
```

### Act

```python
def act(s,i=0):
    action = np.argmax(Q[s,:] + np.random.randn(1,env.action_space.n)/np.square(i/3+1))
    return action
```

### Learn

```python
def learn(s, a, r, s_, a_, done):
    global Q, E
    r_p = Q[s, a] # Predicted reward
    if not done:
        r_ = Q[s_, a_]
        r_t = r + gamma*r_ # Target reward
    else:
        r_t = r
        
    error = r_t - r_p
    
    E[s, :] *= 0
    E[s, a] = 1
    
    Q = Q * LR * error * E  
    E *= gamma*lambda_
    print(Q)
    return Q
```

# Session

```python
for i in range(N_GAMES):
    s = env.reset()
    a = act(s,i)
    for j in range(N_STEPS):
        #env.render()       
        s_, r, done, _ = env.step(a)       
        a_ = act(s_, i)
        _ = learn(s, a, r, s_, a_, done)
        s = s_
        a = a_
        if done:
            break
    
```

```python
rewards = []
for i in range(100):
    s = env.reset()    
    for j in range(N_STEPS):
        
        a  = np.argmax(Q[s])
        s_, r, done, _ = env.step(a)
        #env.render()
        s = s_
        if done:         
            break
    rewards.append(r)
np.mean(rewards)
```
